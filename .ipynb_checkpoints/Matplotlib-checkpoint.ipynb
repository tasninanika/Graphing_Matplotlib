```python
import matplotlib.pyplot as plt
```

```python
x = [2, 3, 4, 8, 9, 10]
y = [5, 6, 7, 11, 12, 13]
```

# Line chart

```python
plt.plot(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# scatter plot

```python
plt.scatter(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Pie chart

```python
x = [1, 2, 3, 4]
plt.pie(x)
plt.show()
```

# Histogram

```python
x = [10000, 15000, 20000, 21000, 30000, 35000, 40000, 45000]
plt.hist(x, bins = [10000, 20000, 30000, 50000])
plt.show()
```

# Bar plot

```python
x = [2, 3, 4, 5, 6, 7]
y = ['F', 'F', 'M', 'M', 'F', 'M']
```

```python
plt.bar(y,x)
plt.show
```

# Box plot

```python
import numpy as np
data = np.random.normal(200,20,300)
data
```

```python
plt.hist(data)
plt.show()
```

```python
np.median(data)
```

```python
plt.boxplot(data)
plt.show()
```

# Violinplot

```python
plt.violinplot(data)
plt.show()
```

# Stackplot

```python
date = [1, 2, 3, 4, 5]
working_hour = [8, 9, 3, 6, 7]
family_time = [3, 2, 5, 4, 6]
plt.stackplot(date, working_hour, family_time)
plt.xlabel('date')
plt.ylabel('time')
plt.show()
```
