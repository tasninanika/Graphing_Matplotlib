```python
import matplotlib.pyplot as plt
```

```python
x = [2, 3, 4, 8, 9, 10]
y = [5, 6, 7, 11, 12, 13]
```

# Line chart

```python
plt.plot(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# scatter plot

```python
plt.scatter(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Pie chart

```python
x = [1, 2, 3, 4]
plt.pie(x)
plt.show()
```

# Histogram

```python
x = [10000, 15000, 20000, 21000, 30000, 35000, 40000, 45000]
plt.hist(x, bins = [10000, 20000, 30000, 50000])
plt.show()
```
