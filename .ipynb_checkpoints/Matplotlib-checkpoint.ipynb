```python
import matplotlib.pyplot as plt
```

```python
x = [2, 3, 4]
y = [5, 6, 7]
```

# Line chart

```python
plt.plot(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# scatter

```python
plt.scatter(x, y)
plt.title('Relationship of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
